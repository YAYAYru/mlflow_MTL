```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import make_pipeline
from sklearn.inspection import DecisionBoundaryDisplay

import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

## Создать двухмерный датасет с бинарной классификацией

```python
NUM_CLASSES = 2
NUM_FEATURES = 2
RANDOM_SEED = 42

X_blob2, y_blob2 = make_blobs(n_samples=100,
    n_features=NUM_FEATURES, 
    centers=NUM_CLASSES, 
    cluster_std=1.5,
    random_state=RANDOM_SEED
)

#plt.figure(figsize=(10, 7))
plt.scatter(X_blob2[:, 0], X_blob2[:, 1], c=y_blob2, cmap=plt.cm.RdYlBu);
```

```python
X, y = X_blob2, y_blob2
X_train2, X_test2, y_train2, y_test2 = train_test_split(
    X, y, test_size=0.4, random_state=42
)
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5

clf2 = KNeighborsClassifier(3)
clf2 = make_pipeline(StandardScaler(), clf2)
clf2.fit(X_train2, y_train2)
```

```python
cm = plt.cm.RdBu
score2 = clf2.score(X_test2, y_test2)
score2
```

```python
cm = plt.cm.RdBu
cm_bright = ListedColormap(["#FF0000", "#0000FF"])


# plt.scatter(X_train2[:, 0], X_train2[:, 1], c=y_train2, cmap=cm_bright, edgecolors="k")
# plt.scatter(
#     X_test2[:, 0], X_test2[:, 1], c=y_test2, cmap=cm_bright, alpha=0.6, edgecolors="k"
# )
DecisionBoundaryDisplay.from_estimator(
    #clf, X, cmap=cm, alpha=0.8, ax=ax, eps=0.5
    clf2, X, cmap=cm, alpha=0.8, eps=0.5, response_method="predict"
)
plt.scatter(
    X_train2[:, 0], X_train2[:, 1], c=y_train2, cmap=cm_bright, edgecolors="k"
)
plt.scatter(
    X_test2[:, 0],
    X_test2[:, 1],
    c=y_test2,
    cmap=cm_bright,
    edgecolors="k",
    alpha=0.6,
)
plt.text(
    x_max - 0.3,
    y_min + 0.3,
    ("%.2f" % score2).lstrip("0"),
    size=15,
    horizontalalignment="right",
    )
plt.tight_layout()
plt.show()
```

## Создать двухмерный датасет с мультиклассной классификацией

```python
cm = plt.cm.RdBu
score3 = clf3.score(X_test3, y_test3)
score3
```

```python
cm = plt.cm.RdBu
cm_bright = ListedColormap(["#FF0000", "#0000FF", "#00FF00"])


# plt.scatter(X_train3[:, 0], X_train3[:, 1], c=y_train3, cmap=cm_bright, edgecolors="k")
# plt.scatter(
#     X_test3[:, 0], X_test3[:, 1], c=y_test3, cmap=cm_bright, alpha=0.6, edgecolors="k"
# )
DecisionBoundaryDisplay.from_estimator(
    #clf, X, cmap=cm, alpha=0.8, ax=ax, eps=0.5
    clf3, X, cmap=cm, alpha=0.8, eps=0.5, response_method="predict"
)
plt.scatter(
    X_train3[:, 0], X_train3[:, 1], c=y_train3, cmap=cm_bright, edgecolors="k"
)
plt.scatter(
    X_test3[:, 0],
    X_test3[:, 1],
    c=y_test3,
    cmap=cm_bright,
    edgecolors="k",
    alpha=0.6,
)
plt.text(
    x_max - 0.3,
    y_min + 0.3,
    ("%.2f" % score3).lstrip("0"),
    size=15,
    horizontalalignment="right",
    )
plt.tight_layout()
plt.show()
```
